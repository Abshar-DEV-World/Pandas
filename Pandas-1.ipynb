#### Merge

```python
# SQL JOIN = Pandas Merge
```

```python
# data is getting generated from two excel files 
# if i wanted to join --> create tables in sql --> insert data --> perform join condition --> extract the output
```

```python
# pandas --> load two excel files directly --> use merge and perform operation 
```

```python
df1 = pd.DataFrame({'id': [1,2,3,4,5], 'stat_marks': [99,97,93,92,98]})
df2 = pd.DataFrame({'id': [3,4,5,6,7], 'ml_marks': [92,94,98,95,89]})
df2
```

```python
# pandas merge = SQL Join
# objective : to extract data for who appeared in both exams
# select * from df1 join df2 on df1.id = df2.id

df1.merge(df2, on = 'id', how = 'inner')
```

```python
df1.merge(df2, on = 'id', how = 'left') # df1 --> left table , df2 --> right table
# NaN --> SQL Null/ empty/ blank/ no value in that place
```

```python
df1.merge(df2, on = 'id', how = 'right')
```

```python
df1.merge(df2, on = 'id', how = 'outer')
```

```python
# Hypothetical Situations # modify data frames accordingly to work
# if the name is different in both data frame ( tables )
# df1.merge(df2, left_on = 'id', right_on = 'sno', how = 'inner') # use left_on , right_on if names are different

# to join on multiple columns
# df1.merge(df2, on = ['id','school'], how = 'inner')
# df1.merge(df2, left_on = ['id','school'], right_on = ['sno','inst'] how = 'inner') # use left_on , right_on if names are different
```

#### Concat

```python
# SQL Union == Pandas Concat
```

```python
df1 = pd.DataFrame({'id': [1,2,3,4,5], 'stat_marks': [99,97,93,92,98]})
df2 = pd.DataFrame({'id': [3,4,5,6,7], 'stat_marks': [92,94,98,95,89]})
# we are stack / join / append one data frame on another
# Ensure column names are same in two data frames
# if objective is to combine monthly data to master data table

pd.concat([df1,df2])
```

```python
df1 = pd.DataFrame({'id': [1,2,3,4,5], 'stat_marks': [99,97,93,92,98]})
df2 = pd.DataFrame({'id': [3,4,5,6,7], 'ml_marks': [92,94,98,95,89]})

pd.concat([df1,df2])
```

```python
# my original data is not getting modified
df1
```

```python
df2
```

```python
df3 = pd.concat([df1,df2])
df3
```

```python
df3.reset_index() # reset index as a funciton is going to store older index as a column by default
# drop = False # by default
```

```python
df3.reset_index(drop = True) # drop = True --> will not save older index
```

```python
# df3 = df3.reset_index(drop = True) execute the operation and store the result in df3
df3.reset_index(drop = True, inplace = True) # execute the operation and store the result in df3
# once change is made permanent you cant roll back
# df4 = df3.reset_index(drop = True) # store in a new variable --> to avoid issues

df3
```
